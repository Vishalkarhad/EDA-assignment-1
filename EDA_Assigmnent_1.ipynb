# **1.What is the distribution of age among heart failure patients in the dataset?**

```python
df['age'].hist(bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Heart Failure Patients')
plt.show()
```

```python
sns.kdeplot(df['age'])
plt.xlabel('Age')
plt.title('KDE of Age Distribution')
plt.show()
```

# **2.How does the death rate vary with age?**

```python
df.groupby('age')['DEATH_EVENT'].mean().plot()
plt.xlabel('Age')
plt.ylabel('Death Rate')
plt.title('Death Rate by Age')
plt.show()
```

# **3.What is the percentage of male and female patients in the dataset?**

```python
df['sex'].value_counts().plot(kind='pie',autopct='%1.1f%%',labels=['male','female'])

plt.title("percentage of male and female in patients")
plt.show()
```

# **4. how does the platelet count vary among different age groups?**

```python
df.groupby('age')['platelets'].mean().plot()
plt.xlabel("AGE")
plt.ylabel("platelet")
plt.title('platelet by age')
plt.show()
```

# **5. is there a correlation between creatinine and sodium levels in the blood?**

```python
df.plot.scatter(x='creatinine_phosphokinase',y='serum_sodium')
plt.xlabel('creatinine_phosphokinase')
plt.ylabel('serum_sodum')
plt.title('correlation between creatinine and sodium level')
plt.show()
```

```python
df[['creatinine_phosphokinase','serum_sodium']].corr()
```

# **6. How does the prevaience of high blood pressure differ between male and female patients?**

```python
df.groupby('sex')['high_blood_pressure'].mean().plot(kind='bar')
plt.xlabel('sex')
plt.ylabel('prevaience of high blood pressure')
plt.title('prevaience of high blood pressure differ between male and female patients')
plt.show()
```

```python
#differ form 0.10 units female has high blood pressure than male
```

# **7. what is the relationship bettwen smoking habits and the occurrence of heart failure?**

```python
df.groupby('smoking')['DEATH_EVENT'].mean().plot(kind='bar')
plt.xlabel('smoking')
plt.ylabel('heart failure')
plt.title('relationship bettwen smoking habits and the occurrence of heart failure')
plt.show()
```

# **8. Are there any noticeable patterns in the distribution of death events across different age groups?**

```python
df.groupby('age')['DEATH_EVENT'].mean().plot()
plt.xlabel('age')
plt.ylabel('death event')
plt.title('death event by age')
plt.show()
```

# **9. Is there any significant difference in ejecion fraction between patients with and without diabetes?**

```python
df.groupby('diabetes')['ejection_fraction'].mean().plot(kind='bar')
plt.xlabel('diabetes')
plt.ylabel('ejection_fraction')
plt.title('difference in ejecion fraction between patients with and without diabetes')
plt.show()
```
